```python
import random

u = 0
c = 0
i = 1
print("Stone, Paper or Scissors?\n")
n = int(input("How many rounds do you want to play? : "))

while i <= n:
    
    User = int(input("Enter 1 for Stone, 2 for Paper or 3 for Scissors: "))
    Computer = random.randint(1,3) 
    print("Computer's Choice: ", Computer, "\n")

    if User == 1 :
        if Computer == 1 :
            print("Stone vs. Stone")
            print("Draw\n")
        elif Computer == 2 :
            print("Stone vs. Paper")
            print("Computer wins!\n")
            c = c + 1
        elif Computer == 3 :
            print("Stone vs. Scissors")
            print("You win!\n")
            u = u + 1
    elif User == 2 :
        if Computer == 1 :
            print("Paper vs. Stone")
            print("You win!\n")
            u = u + 1
        elif Computer == 2 :
            print("Paper vs. Paper")
            print("Draw\n")
        elif Computer == 3 :
            print("Paper vs. Scissors")
            print("Computer wins!\n")
            c = c + 1
    elif User == 3 :
        if Computer == 1 :
            print("Scissors vs. Stone")
            print("Computer wins!\n")
            c = c + 1
        elif Computer == 2 :
            print("Scissors vs. Paper")
            print("You win!\n")
            u = u + 1
        elif Computer == 3 :
            print("Scissors vs. Scissors")
            print("Draw\n")
    else :
        print("Not valid input")
    i = i + 1
    
print("Your Score : ", u, "/", n)
if u > c:
    print("You win!!")
elif u == c:
    print("Draw")
else:
    print("You lose!")
```
